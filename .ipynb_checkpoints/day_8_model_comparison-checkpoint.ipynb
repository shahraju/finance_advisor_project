# Day 8: Model Evaluation & Comparison

In this step, we will compare all the models we trained (Linear Regression, Decision Tree, Random Forest, and Gradient Boosting).  
We’ll create a summary table and also visualize their performance using bar plots for MSE and R².  
This helps us clearly see which model works best for our synthetic stock data.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Collect results from previous days
results = {
    "Linear Regression": {"MSE": 0.13, "R2": 0.99},
    "Decision Tree": {"MSE": 0.25, "R2": 0.98},
    "Random Forest": {"MSE": 0.24, "R2": 0.98},
    "Gradient Boosting": {"MSE": 0.18, "R2": 0.99}
}

# Convert to DataFrame
results_df = pd.DataFrame(results).T
results_df
```

```python
# Bar plot for MSE
plt.figure(figsize=(8,4))
results_df["MSE"].plot(kind="bar", color="skyblue")
plt.title("Model Comparison - MSE")
plt.ylabel("Mean Squared Error")
plt.show()

# Bar plot for R²
plt.figure(figsize=(8,4))
results_df["R2"].plot(kind="bar", color="lightgreen")
plt.title("Model Comparison - R² Score")
plt.ylabel("R² Score")
plt.show()
```

## Observations

From our comparison:

- **Linear Regression**: Very strong performance with low MSE (0.13) and high R² (0.99).  
- **Decision Tree**: Slightly worse, with higher error (MSE = 0.25, R² = 0.98).  
- **Random Forest**: Improved stability but still close to Decision Tree (MSE = 0.24, R² = 0.98).  
- **Gradient Boosting**: Balanced performance, second-best after Linear Regression (MSE = 0.18, R² = 0.99).  

📌 Overall, **Linear Regression** remains the best model for our synthetic dataset, but **Gradient Boosting** is also a strong contender.  
In real-world scenarios, ensemble models like Random Forest and Gradient Boosting may outperform Linear Regression when data becomes more complex or non-linear.
