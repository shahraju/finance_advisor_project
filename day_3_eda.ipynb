# Day 3: Exploratory Data Analysis (EDA)

Today I will explore the synthetic stock dataset that I created earlier.  
The goal is to understand the structure of the data, check if everything looks fine, and get some initial insights before moving to feature engineering.  

I will:  
- Load the dataset and check its shape, columns, and missing values  
- Look at some summary statistics  
- Visualize trends in prices and trading volume  
- Check how stock prices are distributed


```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
stock_data = pd.read_csv("synthetic_stock_data.csv")

# First look at the dataset
print("Shape of dataset:", stock_data.shape)
stock_data.head()
```

```python
# Check for missing values
print("Missing values:\n", stock_data.isnull().sum())
```

```python
# Summary statistics
stock_data.describe()
```

```python
# Plot closing price trend
plt.figure(figsize=(12, 5))
plt.plot(stock_data["date"], stock_data["close"], label="Closing Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Stock Closing Price Over Time")
plt.legend()
plt.show()
```

```python
# Trading volume trend
plt.figure(figsize=(12, 5))
plt.bar(stock_data["date"], stock_data["volume"], color="orange")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.title("Trading Volume Over Time")
plt.show()
```

```python
# Distribution of closing prices
plt.figure(figsize=(8, 5))
sns.histplot(stock_data["close"], kde=True, bins=30, color="blue")
plt.xlabel("Closing Price")
plt.title("Distribution of Closing Prices")
plt.show()
```
